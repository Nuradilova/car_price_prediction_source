Convert to appropriate dtypes

```python
df['Год выпуска'] = df['Год выпуска'].astype('Int64')
df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')
```

```python
# remove NAN columns
df.drop(columns=['Name'], inplace=True)
```

```python
def extract_days(added_at):
    if pd.isna(added_at):  # Handle NaN values
        return None

    # Use regex to find days, months, years, or hours
    match_day = re.search(r"(\d+)\sд\.", added_at)
    match_month = re.search(r"(\d+)\sмес\.", added_at)
    match_year = re.search(r"(\d+)\sгод\.", added_at)
    match_hour = re.search(r"(\d+)\sч\.", added_at)

    if match_day:
        return int(match_day.group(1))
    elif match_month:
        return int(match_month.group(1)) * 30  # Approximate months to 30 days
    elif match_year:
        return int(match_year.group(1)) * 365  # Approximate years to 365 days
    elif match_hour:
        return 0  # Assume hours ago means "today" and returns 0 days
    else:
        return None  # Handle unexpected formats gracefully
```

```python
df["Days Ago"] = df["Added At"].apply(extract_days)
```

```python
df['Upped days'] = df['Upped At'].apply(extract_days)
```

```python
# fill missing Added days and Upped days
days_mean = df['Days Ago'].mean()
fav_mean = df['Upped days'].mean()
df['Days Ago'] = df['Days Ago'].fillna(days_mean)
df['Upped days'] = df['Upped days'].fillna(days_mean)
df['Days Ago'] = df['Days Ago'].astype('int')
df['Upped days'] = df['Upped days'].astype('int')
```

```python
def extract_fuel_type(fuel):
    if pd.isna(fuel):
        return None
    match = re.search(r"/\s*([\w]+)$", fuel)
    if match:
        fuel_type = match.group(1)
        return fuel_type
    

df['Fuel_type'] = df['Двигатель'].apply(extract_fuel_type)
```

```python
def extract_volume(liters):
    if pd.isna(liters):
        return None
    match = re.search(r"\b\d+\.\d+\b", liters) or re.search(r"\b\d+\b", liters)
    if match:        
        return float(match.group(0))
    
df['Volume'] = df['Двигатель'].apply(extract_volume)
```

```python
# some cars fuel type is electro, convert volume to liters
def convert_to_liter(row):
    if row['Fuel_type'] == "электро":  # Check if Fuel Type is "электро"
        specific_output = 80  # 80 horsepower per liter
        return round(row['Volume'] / specific_output, 1)  # Convert Volume to liters
    return row['Volume']  # Return the original Volume for non-electric engines

df['Volume'] = df.apply(convert_to_liter, axis=1)
```

```python
# Extract data from date columns

df["Datetime"] = pd.to_datetime(df["Datetime"])
df["Actual Date"] = df["Datetime"] - pd.to_timedelta(df["Days Ago"], unit="days")
df["Month"] = df["Actual Date"].dt.month_name()

# Function to determine the season
def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    elif month in [9, 10, 11]:
        return "Autumn"

df["Season"] = df["Actual Date"].apply(get_season)
```

```python
# Convert Price columns to int after number extraction
def extract_digits(price):
    cleaned_price = re.sub(r'\D', '', str(price))  # \D removes everything that is not a digit
    return int(cleaned_price) if cleaned_price else None

df['Price_D'] = df['Price (Dollar)'].apply(extract_digits)
df['Price_S'] = df['Price (Som)'].apply(extract_digits)
```

```python
def create_option_columns_multi(data, columns, options):
    """
    Create new columns in the DataFrame for each value in the options list.
    Each column will indicate True/False (1/0) based on whether the value is found
    across multiple specified columns.

    :param data: pd.DataFrame - The input DataFrame.
    :param columns: list - List of column names to check for the presence of values.
    :param options: list - List of distinct values to create columns for.
    :return: pd.DataFrame - The updated DataFrame with new columns.
    """
    for value in options:
        # Check for the value in any of the specified columns
        data[value] = data[columns].apply(lambda row: row.str.contains(value, case=False, na=False)).any(axis=1).astype(int)
    return data

checking_text = [
    "полный электропакет",
    "сигнализация",
    "автозавод",
    "кондиционер",
    "климат контроль",
    "газобалонное оборудование",
    "круиз-контроль",
    "подогрев передних сидений",
    "подогрев всех сидений",
    "обогрев зеркал",
    "ксенон",
    "биксенон",
    "омыватель фар",
    "пневмоподвеска",
    "память сидений",
    "память руля",
    "датчик дождя",
    "датчик света",
    "бортовой компьютер",
    "корректор фар",
    "центральный замок",
    "бесключевой доступ",
    "подогрев руля",
    "вентиляция сидений",
    "лазерные фары",
    "светодиодные фары",  # important features
    "система курсовой устойчивости", 
    "парктроник", 
    "кондиционер", 
    "DVD", 
    "подушки безопасности",
    "кожа",
    "велюр",
    "камера заднего вида",
    "система курсовой устойчивости"
    'тонировка',
    'литые диски',
    'вложений не требует',
    'налог уплачен'

    ]  

df = create_option_columns_multi(df, ['Внешний вид', 'Прочее', 'Безопасность', 'Опции', 'Салон', 'Мультимедиа'], checking_text)
```
