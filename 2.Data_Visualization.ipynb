```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
```

```python
df = pd.read_csv('cleaned_data_25093.csv')
```

```python
categorical_columns = [
    'Кузов', 'Цвет', 'Коробка', 'Привод', 'Руль', 'Состояние', 'Учёт', 'Обмен',
]

# Iterate through the columns and plot bar plots
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    
    # Sort by count to display categories in order of frequency
    order = df[column].value_counts().index
    
    # Create the count plot with a color palette
    sns.countplot(x=column, data=df, order=order, palette='Set2')
    
    # Title and labels
    plt.title(f'Distribution of {column}', fontsize=16, fontweight='bold')
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Count', fontsize=14)
    
    # Rotate the x-ticks for better readability
    plt.xticks(rotation=45, ha='right')
    
    # Add grid lines for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Adjust layout for better spacing
    plt.tight_layout()
    
    # Show the plot
    plt.show()
```

```python
df_int = df[['Views', 'Favorites', 'URL', 'Год выпуска', 'Пробег', 'Кузов', 'Цвет',
       'Коробка', 'Привод', 'Руль', 'Состояние', 'Таможня', 'Наличие',
       'Регион, город', 'Учёт', 'Обмен', 'Рассрочка', 'brand', 'model',
       'restyling', 'Brand_Model_Generation', 'Days Ago', 'Upped days',
       'Fuel_type', 'Volume', 'Actual Date', 'Month', 'Season', 'Price_D',
       'полный электропакет','камера заднего вида', 
       'литые диски', 'вложений не требует', 'налог уплачен']].select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
corr_matrix = df_int.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Integer Features')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 20 most common brands
top_brands = df['brand'].value_counts().head(15)

# Set a modern style
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# Create a barplot with enhanced aesthetics
sns.barplot(
    x=top_brands.index, 
    y=top_brands.values, 
    palette="coolwarm",  # Use a nice color gradient
    edgecolor="black"
)

# Beautify the plot
plt.title('Top 15 Most Common Car Brands', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Brand', fontsize=12, fontweight='bold')
plt.ylabel('Number of Listings', fontsize=12, fontweight='bold')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

# Remove top and right borders for a cleaner look
sns.despine()

# Add value labels on top of bars
for i, value in enumerate(top_brands.values):
    plt.text(i, value + 5, str(value), ha='center', fontsize=10, fontweight='bold', color='black')

plt.tight_layout()
plt.show()
```

```python
# Group by 'brand' and 'Color' to find the average price for each combination
top_brands_by_color = df.groupby(['brand', 'Цвет']).agg({'Price_D': 'mean'}).reset_index()

# Filter to get top 10 brands with highest average price
top_brands_color_sorted = top_brands_by_color.groupby('brand').agg({'Price_D': 'mean'}).sort_values(by='Price_D', ascending=False).head(10)

# Plot the average price by color for top 10 brands
plt.figure(figsize=(14, 8))
sns.barplot(x='brand', y='Price_D', hue='Цвет', data=top_brands_by_color[top_brands_by_color['brand'].isin(top_brands_color_sorted.index)], palette='tab20', dodge=True)
plt.xticks(rotation=45)
plt.title('Top 10 Brands by Average Price and Color')
plt.tight_layout()
plt.legend(title='Color')
plt.show()
```

```python
# Group by 'brand' and 'Год выпуска' to calculate the average price for each combination
top_brands_by_year = df.groupby(['brand', 'Год выпуска']).agg({'Price_D': 'mean'}).reset_index()

# Get the top 10 brands by average price (over all years)
top_brands_sorted = top_brands_by_year.groupby('brand').agg({'Price_D': 'mean'}).sort_values(by='Price_D', ascending=False).head(10)

# Filter data for the top 10 brands
top_brands_by_year_filtered = top_brands_by_year[top_brands_by_year['brand'].isin(top_brands_sorted.index)]

# Plot the average price by year for top 10 brands
plt.figure(figsize=(14, 8))
sns.lineplot(x='Год выпуска', y='Price_D', hue='brand', data=top_brands_by_year_filtered, marker='o', palette='tab10')
plt.title('Top 10 Brands by Average Price Over Years')
plt.xlabel('Year of Manufacture')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Brand')
plt.show()
```

```python
# Group by 'Год выпуска' and 'Fuel_type' to calculate the average price for each combination
top_fuel_by_year = df.groupby(['Год выпуска', 'Fuel_type']).agg({'Price_D': 'mean'}).reset_index()

# Plot the average price by year and fuel type
plt.figure(figsize=(14, 8))
sns.lineplot(x='Год выпуска', y='Price_D', hue='Fuel_type', data=top_fuel_by_year, marker='o', palette='Set2')
plt.title('Average Car Price by Fuel Type and Year')
plt.xlabel('Year of Manufacture')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Fuel Type')
plt.show()
```

```python
# Group by 'brand', 'Год выпуска', and 'Volume' to calculate the average price for each combination
top_brands_by_volume_year = df.groupby(['brand', 'Год выпуска', 'Volume']).agg({'Price_D': 'mean'}).reset_index()

# Filter data for the top 10 brands by average price
top_brands_sorted_volume = top_brands_by_volume_year.groupby('brand').agg({'Price_D': 'mean'}).sort_values(by='Price_D', ascending=False).head(10)
top_brands_by_volume_year_filtered = top_brands_by_volume_year[top_brands_by_volume_year['brand'].isin(top_brands_sorted_volume.index)]

# Plot the average price by volume and year for the top brands
plt.figure(figsize=(14, 8))
sns.lineplot(x='Год выпуска', y='Price_D', hue='brand', data=top_brands_by_volume_year_filtered, marker='o', palette='tab10')
plt.title('Average Price, Fuel Volume, and Year by Brand')
plt.xlabel('Year of Manufacture')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Brand')
plt.show()
```

```python
# Select numerical columns of interest
num_features = ['Price_D', 'Volume', 'Пробег', 'Год выпуска']

# Plot pairplot for relationships
sns.pairplot(df[num_features], diag_kind='kde', plot_kws={'alpha': 0.5})
plt.suptitle('Pairplot of Price, Volume, Mileage, and Year', y=1.02)
plt.tight_layout()
plt.show()
```
